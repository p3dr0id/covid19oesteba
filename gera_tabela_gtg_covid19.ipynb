### Funções utilizadas para agrupar os dados

```
# Agrupa os dados por variável de interesse
def df_agrupado_por_teste(df, nome_var):
  N = df.groupby(nome_var)[nome_var].count()
  perc = N/len(df.index)*100
  result = pd.DataFrame({'N': N, '%': np.round(perc,2)})

  #Acrescenta linha com dados não informados
  ninform = len(df.index)-result['N'].sum() # quantidativo de não informados
  perc_ninform = ninform/len(df.index)*100 # percentual de não informados
  nao_informado = {result.index.name:'Não informado', 'N': ninform, '%': np.round(perc_ninform,2) }

  return result.reset_index().append(nao_informado, ignore_index=True)
```

```
# Agrupa os dados por variável de interesse após renomear categorias sem relevância 
def df_agrupado_por(df, nome_var):

  if (nome_var == 'FAIXA ETARIA'):
     N = df.groupby(nome_var)[nome_var].count()
     perc = N/len(df.index)*100
     result = pd.DataFrame({'N': N, '%': np.round(perc,2)})
     
     #Acrescenta linha com dados não informados
     ninform = len(df.index)-result['N'].sum() # quantidativo de não informados
     perc_ninform = ninform/len(df.index)*100 # percentual de não informados
     nao_informado = {result.index.name:'NAO INFO', 'N': ninform, '%': np.round(perc_ninform,2) }
     return result.reset_index().append(nao_informado, ignore_index=True)

  else:

    df_temp = df.copy()
    df_temp[nome_var] = df_temp[nome_var].replace({r'IGNORAD(O|A)':'NAO INFO', r'INDEFINID(O|A)':'NAO INFO'}, regex=True)
    df_temp[nome_var] = df_temp[nome_var].fillna('NAO INFO')

    N = df_temp.groupby(nome_var)[nome_var].count()
    perc = N/len(df.index)*100
    result = pd.DataFrame({'N': N, '%': np.round(perc,2)})

    #Acrescenta linha com valor total e porcentagem total para conferência
    Ntot = len(df.index) # quantidativo de dados do dataframe original
    perc_tot = result['N'].sum()/Ntot*100 # percentual de não informados
    total = {result.index.name:'TOTAL', 'N': Ntot, '%': np.round(perc_tot,2) }

    return result.reset_index().append(total, ignore_index=True)
```
